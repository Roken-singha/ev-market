```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
ev_adoption_by_year = df['Model Year'].value_counts().sort_index()
plt.figure(figsize=(14,5))
ax = sns.barplot( x=ev_adoption_by_year.index, y= ev_adoption_by_year.values)
ax.bar_label(ax.containers[0], color="red")
plt.xlabel("Model Year")
plt.ylabel("Number of Vehicles Registered")
plt.title("EV growth year wise")
```

```python
# Get top 10 cities with highest EV registrations
top_cities = df['City'].value_counts().head(10).reset_index()
top_cities.columns = ['City', 'Number of Vehicles']  # Rename columns for clarity

# Plotting
plt.figure(figsize=(12, 4))
ax1 = sns.barplot(x='Number of Vehicles', y='City', data=top_cities, palette="crest")
ax1.bar_label(ax1.containers[0], color="red")
plt.title('Top Cities by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
ev_distribution_type = df["Electric Vehicle Type"].value_counts()
plt.figure(figsize=(12, 4))
ax2 = sns.barplot(x=ev_distribution_type.values, y= ev_distribution_type.index)
ax2.bar_label(ax2.containers[0])
plt.xlabel("Vehicle Type")
plt.ylabel("Vehicles Count")
plt.title("Electric Vehicle Type")
```

```python
# Top 10 brands
brand = df["Make"].value_counts().head(10)
plt.figure(figsize=(14,4))
ax3 = sns.barplot(x= brand.index, y = brand.values)
ax3.bar_label(ax3.containers[0])
plt.xlabel("Brand")
plt.ylabel("Count")
plt.title("Top 10 popular brand")
```
